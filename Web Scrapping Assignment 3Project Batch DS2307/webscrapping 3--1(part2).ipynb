```python
#Write a program to scrap all the available details of best gaming laptops from digit.in. 
driver=webdriver.Chrome(r'C:\Users\maham\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe')#WORKING WITH NEW CHROME 116 UPDATE
driver.maximize_window()
```

```python
url="https://www.digit.in/top-products/best-gaming-laptops-under-rs-50-000-3715.html"
driver.get(url)
```

```python
laptop_name=[]
OS=[]
display=[]
processor=[]
memory=[]
price=[]
product_url=[]
```

```python
laptop_name=[]
title_tags=driver.find_elements(By.XPATH,"//span[@class='datahreflink']")
for i in title_tags[0:7]:
    laptop_name.append(i.text)
    
laptop_name
```

```python
details=[]

title_tags=driver.find_elements(By.XPATH,"//div[@class='Specs-Wrap']")
for i in title_tags[0:7]:
    details.append(i.text)
    
laptop_details=[]
for i in range(0,7):
        x=details[i].split("\n")
        laptop_details.append(x)
laptop_details
```

```python
OS=[]
for i in range(0,7):
        x=laptop_details[i][0]
        OS.append(x)
        
        
OS
```

```python
display=[]
for i in range(0,7):
        x=laptop_details[i][1]
        
        display.append(x)
        
display
```

```python
processor=[]
for i in range(0,7):
        x=laptop_details[i][2]
       
        processor.append(x)
        
        
processor
```

```python
memory=[]
for i in range(0,7):
        x=laptop_details[i][3]
       
        memory.append(x)
        
        
memory
```

```python
s=[]

t_tags=driver.find_elements(By.XPATH,"//div[@class='lowest_price']")
for i in t_tags[0:7]:
    s.append(i.text)
    
price=[]
for i in range(0,7):
        d=s[i].split("\n")[0]
        price.append(d)
price
```

```python
product_url=[]
title_tags=driver.find_elements(By.XPATH,"//span[@class='datahreflink']")
for i in title_tags[0:7]:
    l=i.get_attribute("data-href")
    product_url.append(l)
    
product_url
```

```python
import pandas as pd
df=pd.DataFrame({'laptop':laptop_name,'OS':OS,'display':display,'processor':processor,'memory':memory,'processor':processor,'memory':memory,'Price':price,'Product_url':product_url})
df
```
