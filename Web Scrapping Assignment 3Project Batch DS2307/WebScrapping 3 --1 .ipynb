```python
#7  Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: 
#“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”. 
url='https://www.forbes.com/billionaires/'
driver.get(url)
```

```python
ranks = []
names = []
net_worths = []
ages = []
citizenships = []
sources = []
industries = []
```

```python

p=[]
details=[]
title_tags=driver.find_elements(By.XPATH,"//div[@class='TableRow_row__L-0Km TableRow_activeRow__g4oSF']")
for i in title_tags:
    p.append(i.text)
    
list_name=[]

for i in range(0,len(p)):
        x=p[i].split("\n")
        list_name.append(x) 
        
title_tags=driver.find_elements(By.XPATH,"//div[@class='TableRow_row__L-0Km']")
for i in title_tags[0:199]:
    details.append(i.text)

for i in range(0,199):
        x=details[i].split("\n")
        list_name.append(x)
list_name
```

```python
ranks=[]
for i in range(0,200):
        x=list_name[i][0]
        ranks.append(x)
        
        
len(ranks)
```

```python
names=[]
for i in range(0,200):
        x=list_name[i][1]
        names.append(x)
        
        
len(names)
```

```python
net_worths=[]
for i in range(0,200):
        x=list_name[i][2]
        net_worths.append(x)
        
        
net_worths
```

```python
ages=[]

for i in range(0,200):
        x=list_name[i][3]
        ages.append(x)
        
        
ages
```

```python
citizenships=[]

for i in range(0,200):
        x=list_name[i][4]
        citizenships.append(x)
        
        
citizenships
```

```python
sources=[]
for i in range(0,200):
        x=list_name[i][5]
        sources.append(x)
        
        
sources
```

```python
industries=[]

for i in range(0,200):
        x=list_name[i][6]
        industries.append(x)
        
        
industries
```

```python
import pandas as pd
df=pd.DataFrame({'ranks':ranks,'names':names,'net_worths':net_worths,'ages':ages,'citizenships':citizenships,'sources':sources,'industries':industries})
df
```

```python
#8) Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video. 
url="https://youtu.be/RjDhvrQswIo?si=2CTUbrxuVJv2bFqd"
driver.get(url)
driver.maximize_window()
```

```python
comments=[]
comments_upvote=[]
time=[]
```

```python
  
for j in range(0,10):
    title_tags=driver.find_elements(By.XPATH,"//div[@class='style-scope ytd-expander']")
    for i in title_tags[0:80]:
        comments.append(i.text)
    
    for _ in range(5):
        driver.find_element(By.TAG_NAME,'body').send_keys(Keys.END)
    
comments
```

```python
for j in range(0,10):
    
    title_upvotes=driver.find_elements(By.XPATH,"//span[@id='vote-count-middle']")
    
   
    for k in title_upvotes[0:80]:
        comments_upvote.append(k.text)
    
    for _ in range(5):
        driver.find_element(By.TAG_NAME,'body').send_keys(Keys.END)
    
comments_upvote
```

```python
for j in range(0,10):
     title_time=driver.find_elements(By.XPATH,"//yt-formatted-string[@class='published-time-text style-scope ytd-comment-renderer']")
    
    for l in title_time[0:80]:
        time.append(l.text)
    for _ in range(5):
        driver.find_element(By.TAG_NAME,'body').send_keys(Keys.END)
    
time
```

```python
import pandas as pd
df=pd.DataFrame({'Comments':comments,'Upvotes':comments_upvotes,'time':time})
df
```
